```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('robert_downey_jr.jpeg')
```

```python
image_resized = cv2.resize(image, None, fx=1, fy=1)
```

```python
image_cleared = cv2.medianBlur(image_resized, 3)
image_cleared = cv2.medianBlur(image_cleared, 3)
#image_cleared = cv2.medianBlur(image_cleared, 3)
```

```python
image_cleared = cv2.edgePreservingFilter(image_cleared, sigma_s=3)
```

```python
#image_filtered = cv2.bilateralFilter(image_cleared, 3, 10, 5)
image_filtered = image_cleared
for i in range(1):
    image_filtered = cv2.bilateralFilter(image_filtered, 3, 20, 10)

#for i in range(3):
 #   image_filtered = cv2.bilateralFilter(image_filtered, 5, 30, 10)
```

```python
gaussian_mask = cv2.GaussianBlur(image_filtered, (7,7), 2)
image_sharp = cv2.addWeighted(image_filtered, 1.5, gaussian_mask, -0.5, 0)
image_sharp = cv2.addWeighted(image_sharp, 1.4, gaussian_mask, -0.2, 10)
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Original Image
plt.imshow(convertToRGB(image_resized))
```

```python
#Clear Impurities
plt.imshow(convertToRGB(image_cleared))
```

```python
#Final Image
plt.imshow(convertToRGB(image_sharp))
```
